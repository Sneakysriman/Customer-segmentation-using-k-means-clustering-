```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans


 


print(df.head())
print(df.info())

# Step 3: Optional - Rename columns for easy access
df.columns = ['CustomerID', 'Gender', 'Age', 'Annual Income', 'Spending Score']

# Step 4: Select features for clustering
X = df[['Annual Income', 'Spending Score']]

# Step 5: Apply K-means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# Step 6: Visualize the clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Annual Income', y='Spending Score', hue='Cluster', palette='Set2', s=100)
plt.title('Customer Segments')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()

# Step 7: Optional - Save segmented data
df.to_csv('segmented_customers.csv', index=False)
```
